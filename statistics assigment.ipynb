Write a function Goal_diff_count to calculate and return goal differences

```python
#function to calculate the goal differnce 
def Goal_diff_count(df):
    #calclate the goal differnce
    df['Goal Difference'] = df['GoalsFor'] - df['GoalsAgainst']
    return df[['Team', 'Goal Difference']]
```
